```python
wetness = np.linspace(0, 100, 100)
precipitation_deposits = np.linspace(0, 100, 100)
ice_thickness = np.linspace(0, 100, 100)
```

```python
plt.figure(figsize=(4, 4), dpi=1000)

plt.title('Reduced Friction in Rainy Weather')
plt.xlim([0, 100])
plt.ylim([0, 1])

plt.plot(np.exp(-0.916*wetness/100) * (1-wetness/100)**3 * 0.6 + 0.4, color='blue', label='Friction with Wetness')
plt.plot(1-0.1*precipitation_deposits/100, color='black', label='Friction with Precipitation Deposits')
plt.legend(loc="lower right")
plt.xlabel('Intensity of wetness and precipitation deposit')
plt.ylabel('Measured Friction')
```

```python
plt.figure(figsize=(4, 4), dpi=1000)

plt.title('Reduced Friction in Icy Weather')
plt.xlim([0, 100])
plt.ylim([0, 1])


plt.plot(np.exp(-1.89711*ice_thickness/100) * (1-ice_thickness/100)**3 * 0.85 + 0.15, color='blue', label='Friction with Ice Thickness')
plt.legend(loc="lower right")

plt.xlabel('Ice Thickness')
plt.ylabel('Measured Friction')
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f(wetness, precipitation_deposits):
    return np.exp(-0.916*wetness/100) * (1-wetness/100)**3 * 0.6 + 0.4 - 0.1 * precipitation_deposits/100

x = np.linspace(0, 100, 100)
y = np.linspace(0, 100, 100)
wetness, precipitation_deposits = np.meshgrid(x, y)

Z = f(wetness,precipitation_deposits)
fig = plt.figure(figsize = (13,13), dpi=200)
ax = plt.axes(projection='3d')

ax.plot_surface(wetness, precipitation_deposits, Z, cmap='jet', edgecolor='None', rstride=1, cstride=1)
ax.set_title("Reduced Friction in Rainy Weather", fontsize = 18)
ax.set_xlabel('Intensity of Wetness', fontsize = 11)
ax.set_ylabel('Intensity of Precipitation Deposits', fontsize = 11)
ax.set_zlabel('Measured Friction', fontsize = 11)
```
